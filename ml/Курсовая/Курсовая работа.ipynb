Создадим функцию для предобработки данных

```python
class preprocessor():
    def fit_X(self, X):
        X['LifeSquare_isna'] = 0
        X.loc[df['LifeSquare'].isna(), 'LifeSquare_isna'] = 1
        X.drop('Healthcare_1', axis=1, inplace=True)
        X_temp = X[X['LifeSquare'].notna()][['Square', 'LifeSquare']]
        avdep_life = X_temp['Square'] / X_temp['LifeSquare']
        avdep_life_mean = avdep_life.mean()
        self.avdep_life_mean = avdep_life_mean
        indexes = X[X['LifeSquare'] > X['Square']][['Square', 'LifeSquare']].index
        for i in indexes:
            X.loc[i,'LifeSquare'] = X.loc[i,'Square'] / avdep_life_mean
        indexes = X.loc[X['LifeSquare'].isna()].index
        for i in indexes:
            X.loc[i,'LifeSquare'] = X.loc[i,'Square'] / avdep_life_mean


        X_temp = X[['Square','KitchenSquare']]
        avdep_kitch = X_temp['Square'] / X_temp['KitchenSquare']
        avdep_kitch_mean = avdep_kitch.mean()
        self.avdep_kitch_mean = avdep_kitch_mean
        indexes = X[X['KitchenSquare'] > X['Square']][['Square', 'KitchenSquare']].index
        for i in indexes:
            X.loc[i,'KitchenSquare'] = X.loc[i,'Square'] / avdep_kitch_mean   
        X = pd.concat([X, pd.get_dummies(X[['Ecology_2', 'Ecology_3', 'Shops_2']])], axis=1)
        X.drop(columns=['Id','Ecology_2', 'Ecology_3', 'Shops_2'] , inplace=True)

        return X
    
    def fit_y(self, y):

        y['LifeSquare_isna'] = 0
        y.loc[df['LifeSquare'].isna(), 'LifeSquare_isna'] = 1
        y.drop('Healthcare_1', axis=1, inplace=True)
        indexes = y[y['LifeSquare'] > y['Square']][['Square', 'LifeSquare']].index
        for i in indexes:
            y.loc[i,'LifeSquare'] = y.loc[i,'Square'] / self.avdep_life_mean
        indexes = y.loc[y['LifeSquare'].isna()].index
        for i in indexes:
            y.loc[i,'LifeSquare'] = y.loc[i,'Square'] / self.avdep_life_mean


        y_temp = y[['Square','KitchenSquare']]

        indexes = y[y['KitchenSquare'] > y['Square']][['Square', 'KitchenSquare']].index
        for i in indexes:
            y.loc[i,'KitchenSquare'] = y.loc[i,'Square'] / self.avdep_kitch_mean


        y = pd.concat([y, pd.get_dummies(y[['Ecology_2', 'Ecology_3', 'Shops_2']])], axis=1)
        y.drop(columns=['Id','Ecology_2', 'Ecology_3', 'Shops_2'] , inplace=True)
        return y
```
